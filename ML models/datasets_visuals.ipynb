# GDP

```python
sns.scatterplot(data=gdp, x="Country Name", y="Gdp")
sns.set(rc={'figure.figsize':(20,10)})
```

```python
sns.lineplot(data=gdp, x="Country Name", y="Gdp")
sns.set(rc={'figure.figsize':(20,10)})
```

```python
sns.barplot(data=gdp, x="Country Name", y="Gdp")
sns.set(rc={'figure.figsize':(20,10)})
```

```python
Top_5 = gdp.groupby("Country Name").Gdp.sum().nlargest(n =5)
```

```python
Top_5
```

```python
plt.style.use('seaborn')
Top_5.plot(kind ='bar', figsize =(8,8), fontsize =14,color= 'darkgreen')


plt.show()
```

```python
bottom_5 = gdp.groupby("Country Name").Gdp.sum().nsmallest(n =5)
```

```python
bottom_5
```

```python
plt.style.use('seaborn')
bottom_5.plot(kind ='bar', figsize =(8,8), fontsize =14,color= 'maroon')


plt.show()
```
